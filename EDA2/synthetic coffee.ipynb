<h2>indroduction</h2>

The Synthetic Coffee Health Dataset provides detailed information about coffee consumption and its connection with health and lifestyle. It includes factors such as age, gender, country, coffee intake, caffeine levels, sleep hours, sleep quality, BMI, heart rate, stress levels, physical activity, and personal habits like smoking and alcohol use. With these attributes, the dataset gives a broad view of how coffee habits vary across individuals.

By studying this dataset, it is possible to explore how caffeine intake may influence sleep quality, stress, or physical activity. It also highlights health-related aspects such as body mass index, heart rate, and the presence of health issues, allowing patterns to be observed among different groups of people.

In addition, the dataset allows comparisons between demographic categories such as age, gender, and country, helping to identify differences in coffee consumption and its effects. This makes it useful for gaining insights into lifestyle trends and the potential role of coffee in everyday health and well-being.

<h3>Data Understanding</h3>

```python
# Basic information about the dataset
print(df.info())
```

```python
# Coffee intake: Male vs Female
print(df.groupby("Gender")["Coffee_Intake"].mean())
sns.barplot(x="Gender", y="Coffee_Intake", data=df, estimator="mean", palette="Set2")
plt.title("Average Coffee Intake by Gender")
plt.show()
```

```python
#Sleep hours: Good vs Poor quality
print(df.groupby("Sleep_Quality")["Sleep_Hours"].mean())
sns.boxplot(x="Sleep_Quality", y="Sleep_Hours", data=df, palette="pastel")
plt.title("Sleep Hours by Sleep Quality")
plt.show()
```

```python
# Caffeine intake: Stress levels
print(df.groupby("Stress_Level")["Caffeine_mg"].mean())
sns.barplot(x="Stress_Level", y="Caffeine_mg", data=df, estimator="mean", palette="coolwarm")
plt.title("Caffeine Intake by Stress Level")
plt.show()
```

```python
#BMI: Smokers vs Non-smokers
print(df.groupby("Smoking")["BMI"].mean())
sns.boxplot(x="Smoking", y="BMI", data=df, palette="Set3")
plt.title("BMI of Smokers vs Non-Smokers")
plt.show()
```

```python
#Heart rate: Alcohol users vs Non-users
print(df.groupby("Alcohol_Consumption")["Heart_Rate"].mean())
sns.barplot(x="Alcohol_Consumption", y="Heart_Rate", data=df, estimator="mean", palette="muted")
plt.title("Heart Rate: Alcohol Users vs Non-Users")
plt.show()
```

```python
#Physical activity: With vs Without health issues
print(df.groupby("Health_Issues")["Physical_Activity_Hours"].mean())
sns.barplot(x="Health_Issues", y="Physical_Activity_Hours", data=df, estimator="mean", palette="Accent")
plt.title("Physical Activity Hours by Health Issues")
plt.xticks(rotation=30)
plt.show()
```

```python
#Coffee intake: Different occupations
print(df.groupby("Occupation")["Coffee_Intake"].mean())
sns.barplot(x="Occupation", y="Coffee_Intake", data=df, estimator="mean", palette="husl")
plt.title("Coffee Intake by Occupation")
plt.xticks(rotation=30)
plt.show()
```

```python
#Sleep hours: Low, Medium, High caffeine
df['Caffeine_Level'] = pd.cut(df['Caffeine_mg'], bins=[0,100,300,600], labels=["Low","Medium","High"])
print(df.groupby("Caffeine_Level")["Sleep_Hours"].mean())
sns.boxplot(x="Caffeine_Level", y="Sleep_Hours", data=df, palette="Set1")
plt.title("Sleep Hours by Caffeine Level")
plt.show()
```

```python
#Caffeine intake: Age groups
df['Age_Group'] = pd.cut(df['Age'], bins=[0,30,50,100], labels=["<30","30-50",">50"])
print(df.groupby("Age_Group")["Caffeine_mg"].mean())
sns.barplot(x="Age_Group", y="Caffeine_mg", data=df, estimator="mean", palette="Paired")
plt.title("Caffeine Intake by Age Group")
plt.show()
```

```python
#Stress level distribution: Countries
stress_country = pd.crosstab(df["Country"], df["Stress_Level"], normalize="index") * 100
print(stress_country.head())
stress_country.plot(kind="bar", stacked=True, figsize=(10,5), colormap="tab20")
plt.title("Stress Level Distribution by Country (%)")
plt.ylabel("Percentage")
plt.show()
```

```python
#Sleep hours: Smokers vs Non-smokers
print(df.groupby("Smoking")["Sleep_Hours"].mean())
sns.boxplot(x="Smoking", y="Sleep_Hours", data=df, palette="Set2")
plt.title("Sleep Hours of Smokers vs Non-Smokers")
plt.show()
```

```python
# Distribution of Coffee Intake
sns.histplot(df["Coffee_Intake"], bins=30, kde=True, color="brown")
plt.title("Distribution of Coffee Intake")
plt.show()
```

```python
#Age distribution
sns.histplot(df["Age"], bins=30, kde=True, color="teal")
plt.title("Age Distribution")
plt.show()
```

```python
#Avg Coffee Intake: Male vs Female
sns.barplot(x="Gender", y="Coffee_Intake", data=df, estimator="mean", palette="Set2")
plt.title("Average Coffee Intake by Gender")
plt.show()
```

```python
#Scatter: Caffeine vs Sleep Hours, colored by Stress Level
sns.scatterplot(x="Caffeine_mg", y="Sleep_Hours", hue="Stress_Level", data=df, palette="Set1", alpha=0.6)
plt.title("Caffeine vs Sleep Hours by Stress Level")
plt.show()
```

```python
#Avg Coffee Intake by Occupation
sns.barplot(x="Occupation", y="Coffee_Intake", data=df, estimator="mean", palette="husl")
plt.title("Average Coffee Intake by Occupation")
plt.xticks(rotation=30)
plt.show()
```

<h2>suggestions and Recommendations</h2>

1. Coffee & Caffeine Consumption

Average coffee intake is moderate, but a significant group consumes high caffeine (>300 mg daily).

Recommendation: People with high caffeine intake should reduce consumption, as excess caffeine can lead to poor sleep quality and increased stress.

2. Sleep Patterns

Many individuals with high coffee consumption show reduced sleep hours and poorer sleep quality.

Recommendation: Avoid drinking coffee late in the evening to improve sleep. Promoting healthier bedtime habits could improve overall well-being.

3. Stress Levels

Higher caffeine consumers often show medium to high stress levels.

Recommendation: Encourage stress management practices (exercise, meditation) and reduce excessive caffeine intake.

4. Physical Activity

People with low stress levels generally have more physical activity hours.

Recommendation: Increasing daily physical activity (walking, exercise) can reduce stress and balance the effects of caffeine.

5. Health Issues

Mild health issues (e.g., related to BMI or sleep) are more common in groups with higher caffeine intake.

Recommendation: Regular health check-ups and monitoring BMI and heart rate can help in early detection of problems.

6. Smoking & Alcohol Habits

Smokers and alcohol users tend to show slightly higher caffeine intake and more health risks.

Recommendation: Awareness campaigns can be designed to reduce unhealthy combinations of caffeine, smoking, and alcohol.

7. Demographic Insights

Coffee intake varies across age groups, countries, and occupations. Younger groups and office workers consume more coffee.

Recommendation: Workplace wellness programs can provide healthier alternatives to coffee for energy and focus.
